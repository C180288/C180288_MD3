### 描画する関数は、

$$f(x) = x ^ 2 + 2x + 1 $$

### という二次関数です。
### Jupyter Notebook中にグラフを描画するため、いつも通りにいくつかのモジュールをインポートして準備しましょう。

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

### 【設問２－１】解答
#### 二次関数f(x)を定義してください。関数名は、function_1(x)としてください。

```python
def function_1(x):
    return x**2 + 2*x + 1
  
```

```python
### 【設問２－２】解答
#### 関数f(x)におけるxは、-10から+10までの値を取り得るものとします。このxの範囲をnp.arange(下限値, 上限値)を使って定義してください。
#### arrangeではなく、arangeであることに注意してください。
x = np.arange(-10,11)
x
```

### では、関数のグラフを描画しましょう。

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(10, 4))
plt.plot(x, function_1(x))
plt.grid(True)
```

## 【設問３】
### Pandasを使ってcsvファイルを読み込み、統計量を計算しましょう。
### 読み込むファイルは、titanic.csvです。

### 【設問３－１】解答
#### titanic.csvをDataFrame型データに読み込み、先頭の５行を表示させてください。

```python
titanic = pd.read_csv('titanic.csv')
titanic.loc[0:4]
```

### 【設問３－２】解答
#### Titanic号の乗員の年齢構成を調べ、
+ 9歳以下　　 　〇〇人
+ 10歳から19歳　〇〇人
* 20歳から29歳　〇〇人
* 30歳から39歳　〇〇人
+ 40歳から49歳　〇〇人
+ 50歳から59歳　〇〇人
+ 60歳から69歳　〇〇人
+ 70歳以上　　　〇〇人
#### のように表示しなさい。

```python
ageCount = pd.Series(
    [titanic[titanic['Age']<10].PassengerId.count(),
    titanic[titanic['Age'].isin(np.arange(10,20))].PassengerId.count(),
    titanic[titanic['Age'].isin(np.arange(20,30))].PassengerId.count(),
    titanic[titanic['Age'].isin(np.arange(30,40))].PassengerId.count(),
    titanic[titanic['Age'].isin(np.arange(40,50))].PassengerId.count(),
    titanic[titanic['Age'].isin(np.arange(50,60))].PassengerId.count(),
    titanic[titanic['Age'].isin(np.arange(60,70))].PassengerId.count(),
    titanic[titanic['Age']>=70].PassengerId.count()],
    index=('・9歳以下', 
           '・10歳から19歳', 
           '・20歳から29歳',
           '・30歳から39歳', 
           '・40歳から49歳',
           '・50歳から59歳', 
           '・60歳から69歳', 
           '・70歳以上')
)
print(ageCount.apply(lambda x: str(x)+"人"))
```

### 【設問３－３】解答
#### 設問３－２で求めた年齢構成を円グラフで表してください。


```python
index = ("Age<10",
        "Age:10~19",
        "Age:20~29",
        "Age:30~39",
        "Age:40~49",
        "Age:50~59",
        "Age:60~69",
        "Age>70",)
plt.pie(ageCount, labels=index, autopct="%.1f%%")
plt.title("Titanic's age distribution")
plt.show()
```
